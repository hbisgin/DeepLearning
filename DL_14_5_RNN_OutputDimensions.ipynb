```python
import torch
import torch.nn as nn

# ============================================
# 1. Define dimensions
# ============================================

B = 2   # batch size (2 sequences)
T = 3   # sequence length
D = 4   # input size
H = 5   # hidden size

# ============================================
# 2. Create dummy input
# ============================================

x = torch.randn(B, T, D)
print(x)

print("Input shape (B, T, D):", x.shape)


# ============================================
# 3. Define RNN
# ============================================

rnn = nn.RNN(input_size=D, hidden_size=H, batch_first=True)


# ============================================
# 4. Forward pass
# ============================================

output, h_n = rnn(x)

print("\nOutput shape (B, T, H):", output.shape)
print("Final hidden shape (1, B, H):", h_n.shape)
print(output)
print(h_n)
```

```python
W_input = rnn.weight_ih_l0   # (H, D)
W_hidden = rnn.weight_hh_l0  # (H, H)

print(W_input)
print(W_hidden)
```
